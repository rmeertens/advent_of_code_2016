```python
import numpy as np
from collections import defaultdict

def run_code(init_value): 
    lines= open('inputs/day25.txt').readlines()
    lines = [line.strip() for line in lines]
    
    registers = defaultdict(int)
    registers['a'] = init_value
    index = 0

    pattern_out = ""

    visited = set()
    while index < len(lines) : 
        state = str(index) + str(registers) + str(lines)
        if state in visited: 
            break # now it repeats
        visited.add(state)
        
        command, *arguments = lines[index].split(' ')
        
        if command == 'tgl': 
            try: 
                a = int(arguments[0])
                changeindex = index + a
            except: 
                changeindex = index + registers[arguments[0]]

            if changeindex >= len(lines): 
                pass
            elif 'inc' in lines[changeindex]: 
                lines[changeindex] = lines[changeindex].replace('inc', 'dec')
            elif 'dec' in lines[changeindex]: 
                lines[changeindex] = lines[changeindex].replace('dec', 'inc')
            elif 'tgl' in lines[changeindex]: 
                lines[changeindex] = lines[changeindex].replace('tgl', 'inc')
            elif 'out' in lines[changeindex]: 
                lines[changeindex] = lines[changeindex].replace('out', 'inc')

                
            # two arguments
            elif 'jnz' in lines[changeindex]: 
                lines[changeindex] = lines[changeindex].replace('jnz', 'cpy')
            elif 'cpy' in lines[changeindex]: 
                lines[changeindex] = lines[changeindex].replace('cpy', 'jnz')
            else: 
                print("WTF", lines[changeindex])

        if command == "cpy": 
            a, b = arguments
            try: 
                a = int(a)
                registers[b] = a
            except: 
                registers[b] = registers[a]

        if command == "inc": 
            a = arguments[0]
            registers[a] += 1
        if command == "dec": 
            a = arguments[0]
            registers[a] -= 1
        if command == "jnz": 
            a, b = arguments
            try: 
                a = int(a)
            except: 
                a = registers[a]

            try:
                b = int(b)
            except: 
                b = registers[b]

            if a != 0: 
                index += b -1
        if command == "out": 
            try: 
                a = int(arguments[0])
            except: 
                a = registers[arguments[0]]
            pattern_out += str(a)
        index += 1
    return pattern_out

# for i in range(1000000): 

start = 2
while True: 
    start+=1
    out = run_code(start)
    print(start)
    
    out = out+out # has to repeat
    found = True
    for i in range(len(out)-1): 
        if out[i] == out[i+1]: 
            found = False
            break
    if found: 
        print("FOUND IT", start)
        break
```
