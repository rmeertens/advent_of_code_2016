```python
import re
import matplotlib.pyplot as plt
import numpy as np
lines = [[x for x in line.strip().split(" ") if len(x) > 0] for line in open('inputs/day22.txt')]
nodes = lines[2:]
lines[1]
```

```python
viable_pairs = 0

for name1, size1, used1, available1, percentage1 in nodes: 
    used_a = int(used1[:-1])
    if used_a == 0: 
        # Node A is not empty (its Used is not zero).
        continue
    
    for name2, size2, used2, available2, percentage2 in nodes: 
        if name1==name2: 
            # Nodes A and B are not the same node.
            continue
        avail_b = int(available2[:-1])
        if used_a <= avail_b: 
            viable_pairs += 1

print("Part 1", viable_pairs)
```

```python
nodes.sort()

maxx, maxy = 0, 0
for name1, size1, used1, available1, percentage1 in nodes:
    ints = [int(a) for a in re.findall(r'\d+', name1)]
    maxx = max(maxx, ints[0])
    maxy = max(maxy, ints[1])
maxx, maxy
```

```python
# Note: this is a global state!
sizes = np.zeros((maxy+1, maxx+1)) # Y, X

useds = np.zeros((maxy+1, maxx+1)) # Y, X
goal_pos = np.zeros((maxy+1, maxx+1)) # Y, X

for name1, size1, used1, available1, percentage1 in nodes:
    ints = [int(a) for a in re.findall(r'\d+', name1)]
    sizes[ints[1], ints[0]] = int(size1[:-1])
    useds[ints[1], ints[0]] = int(used1[:-1])
    
plt.imshow(sizes)
plt.show()
plt.imshow(useds)
plt.show()
plt.imshow(useds==0)
plt.show()
```

```python
emtpy_y, empty_x = np.where(useds==0)
emtpy_y, empty_x = emtpy_y[0], empty_x[0]
large_line_index = np.argmax(np.sum(sizes, axis=1))
emtpy_y, empty_x
```

```python
from dataclasses import dataclass
import numpy as np
import copy

@dataclass
class ComputerState:
    current_filled: np.array
    goal_square: (int,int)
    empty_square: (int,int)

    def heuristic(self) -> float:
        if self.empty_square[1] >= large_line_index: 
            return self.empty_square[0]+100000
        elif self.empty_square[1] >= 3: 
            return self.empty_square[1]+10000
        else: 
            return abs(self.goal_square[0]-self.empty_square[0])* + abs(self.goal_square[1]-self.empty_square[1]) + self.goal_square[0]*20 + self.goal_square[1]*20
#             return abs(self.goal_square[0]-self.empty_square[0])/2 + abs(self.goal_square[1]-self.empty_square[1])/2
    def representation(self) -> str: 
        return str(self.current_filled[:2,:]) + str((self.current_filled > 0.0)[2:, :])

def expand(state): 
    """Return all the possible new boards, and their evaluation of our heuristic"""
    x,y = state.empty_square
    maxy, maxx = state.current_filled.shape
    
    newstates = list()
    
    for dx, dy in [(-1, 0), (1,0), (0,-1), (0,1)]: 
        nx, ny = x+dx, y+dy
        
        if 0 <= nx < maxx and 0 <= ny < maxy: # viable new location for the empty square!
                # The data also has to be transferable!
                if sizes[y,x] < state.current_filled[ny,nx]: 
                    continue
                    
                # Create a new copy of our state
                newstate = copy.deepcopy(state)
                newstate.current_filled = state.current_filled.copy()
                                
                # swap the empty state with the new state on the board
                newstate.current_filled[y,x] = newstate.current_filled[ny,nx]
                newstate.current_filled[ny,nx] = 0.0

                # set the location of the empty square 
                newstate.empty_square = (nx, ny)
                # if the empty square is moving the goal square: adjust this!
                if (nx,ny) == state.goal_square: 
#                     print("Moving goal square")
                    newstate.goal_square = (x,y) 
                newstates.append(newstate)
    return newstates

a = ComputerState(current_filled=useds, goal_square=(maxx,0), empty_square=(empty_x, emtpy_y))

# A* search
import heapq
from itertools import count

# a global
tiebreaker = count()

expand_heap = list()
maxval = 1000000
heapq.heappush(expand_heap, (0+a.heuristic(), 0, 0, a))

visited = set()

maxsteps = 0
front = 0
while expand_heap: 
    _, steps_so_far, _, state = heapq.heappop(expand_heap)
#     print(steps_so_far)
    front += 1
    
    if state.goal_square == (0,0): 
        print("PART 2 REACHED IN", steps_so_far)
        break
    if steps_so_far > maxsteps: 
        print(steps_so_far)
        maxsteps = steps_so_far
        plt.imshow(state.current_filled)
        plt.show()

        print("^^", state.goal_square, 'front', front)
        front = 0
        
    for newstate in expand(state): 
        if newstate.representation() in visited: 
            continue
        visited.add(newstate.representation())
        
        newg = steps_so_far+1+newstate.heuristic()
#         print(newstate)
        heapq.heappush(expand_heap, (newg, steps_so_far+1, next(tiebreaker), newstate))
```
